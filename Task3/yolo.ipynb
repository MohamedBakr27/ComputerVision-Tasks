```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
model = YOLO('../yolov8n.pt')
results = model(source='img.jpg', show=True, conf=0.4, save=True)
```

```python
img = results[0].plot()
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
plt.show()
```
